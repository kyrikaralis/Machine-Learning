```python
#Logitical Regression
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
#read in the CSV
data = pd.read_csv(os.path.join("cleaned_census_1.csv"))
data.head()
```

```python
#try this with other variables:
# Assign X (data) and y (target)
X = data.drop("education_level", axis=1)
y = data["marital-status"]
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
#print training score
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

```python
#make predictions
predictions = classifier.predict(X_test)
print(f"First 10 Predictions:   {predictions[:10]}")
print(f"First 10 Actual labels: {y_test[:10].tolist()}")
```

```python
 pd.DataFrame({"Prediction": predictions, "Actual": y_test}).reset_index(drop=True)
```

```python
import pickle

#exporting this model via pickle so we can use it on the flask app
pickle.dump(classifier, open("model.pkl","wb"))
```
